a. Obtain the 50th Fibonacci number

```
.data
output: .asciiz "The 50th Fibonacci number is: "

.text
.globl main
main:
    # Calculate the 50th Fibonacci number
    li $t0, 0           # First term
    li $t1, 1           # Second term
    li $t2, 2           # Counter
    
.loop:
    add $t3, $t0, $t1    # Add terms
    move $t0, $t1        # Swap terms
    move $t1, $t3
    addi $t2, $t2, 1     # Increment counter
    bne $t2, 50, .loop   # If counter is not equal to 50, jump to loop
    
    # Output the result
    li $v0, 4
    la $a0, output
    syscall
    li $v0, 1
    move $a0, $t1        # Output the 50th Fibonacci number
    syscall
    
    # Exit the program
    li $v0, 10
    syscall
```

b. Find out if a given year is leap

```
.data
prompt: .asciiz "Enter a year: "
output_leap: .asciiz "The year is a leap year."
output_not_leap: .asciiz "The year is not a leap year."

.text
.globl main
main:
    # Prompt user for input
    li $v0, 4
    la $a0, prompt
    syscall
    
    # Read user input
    li $v0, 5
    syscall
    
    # Save user input in $t0
    move $t0, $v0
    
    # Check if the year is a leap year
    andi $t1, $t0, 3        # Check if the year is divisible by 4
    bnez $t1, not_leap      # If not divisible by 4, it's not a leap year
    
    andi $t1, $t0, 100      # Check if the year is divisible by 100
    bnez $t1, century       # If divisible by 100, go to century check
    
    andi $t1, $t0, 400      # Check if the year is divisible by 400
    beqz $t1, not_leap      # If not divisible by 400, it's not a leap year
    
    j leap                  # It's a leap year
    
century:
    andi $t1, $t0, 400      # Check if the year is divisible by 400
    bnez $t1, leap          # If divisible by 400, it's a leap year
    
not_leap:
    # Output that it's not a leap year
    li $v0, 4
    la $a0, output_not_leap
    syscall
    
    j exit
    
leap:
    # Output that it's a leap year
    li $v0, 4
    la $a0, output_leap
    syscall
    
exit:
    # Exit the program
    li $v0, 10
    syscall
```

c. Given an array of integers, calculate the average

```
.data
array: .word 10, 20, 30, 40, 50     # Example array of integers
array_size: .word 5                  # Size of the array
output: .asciiz "The average is: "

.text
.globl main
main:
    # Calculate the average of the array
    la $t0, array            # Load the base address of the array
    lw $t1, array_size       # Load the size of the array
    li $t2, 0               # Initialize sum to 0
    
    # Loop through the array and calculate the sum
    li $t3, 0               # Counter
.loop:
    beq $t3, $t1, .calc_avg   # If counter equals array size, jump to calculate average
    
    add $t2, $t2, 0($t0)    # Add the current element to the sum
    addi $t0, $t0, 4        # Move to the next element
    addi $t3, $t3, 1        # Increment the counter
    
    j .loop                 # Jump to loop
    
.calc_avg:
    # Calculate the average
    div $t2, $t1            # Divide the sum by the size of the array
    mflo $t2                # Move the quotient to $t2 (the average)
    
    # Output the result
    li $v0, 4
    la $a0, output
    syscall
    
    li $v0, 1
    move $a0, $t2           # Output the average
    syscall
    
    # Exit the program
    li $v0, 10
    syscall
```

d. Write a program to transform from Celsius to Fahrenheit and vice versa

```
.data
prompt: .asciiz "Enter the temperature: "
output_celsius: .asciiz "Temperature in Celsius: "
output_fahrenheit: .asciiz "Temperature in Fahrenheit: "

.text
.globl main
main:
    # Prompt user for input
    li $v0, 4
    la $a0, prompt
    syscall
    
    # Read user input
    li $v0, 5
    syscall
    
    # Save user input in $t0
    move $t0, $v0
    
    # Convert Celsius to Fahrenheit
    mul $t1, $t0, 9       # Multiply Celsius by 9
    div $t1, $t1, 5       # Divide by 5
    addi $t1, $t1, 32     # Add 32 to the result
    
    # Output temperature in Celsius
    li $v0, 4
    la $a0, output_celsius
    syscall
    li $v0, 1
    move $a0, $t0
    syscall
    
    # Output temperature in Fahrenheit
    li $v0, 4
    la $a0, output_fahrenheit
    syscall
    li $v0, 1
    move $a0, $t1
    syscall
    
    # Exit the program
    li $v0, 10
    syscall
```

### 4. What are the differences among sequential access, direct access, and random access?

Todos estos tipos de acceso, son metodos para acceder a la informacion que se encuentra en la memoria. La diferencia entre ellos es la forma en la que se accede a la informacion. Por un lado el metodo sequential access tiene la memoria organizada en unidades de datos llamados registros, para poder acceder a estos datos se debe de seguir una secuencia lineal, esto lo que hace es que se vaya por cada registro hasta llegar al que se desea acceder. Por otro lado, el metodo direct access, tiene la memoria organizada en bloques de datos, y para acceder a estos bloques de datos se debe de conocer la direccion de memoria en la que se encuentra el bloque de datos que se desea acceder. Finalmente, el metodo random access, tiene la memoria organizada en un mecanismo de direccionamiento, este tiene un unico mecanismo de direccionamiento porque es fisicamenete conectado a la memoria, en el cual se puede acceder a cualquier direccion de memoria en cualquier momento, sin necesidad de seguir una secuencia lineal o conocer la direccion de memoria en la que se encuentra el bloque de datos que se desea acceder.

Estos 3 metodos son bastante diferentes y algunos mas lentos que otros, sin embargo, al tener un acceso a memoria tan distinto, permite que cada uno de ellos sea utilizado para distintos tipos de almacenamiento de memoria y para distintos patrones de acceso a memoria.

### 5. What common characteristics are shared by all RAID levels?

Las caracteristicas que comparten todos los niveles de RAID son las siguientes:

Multiples Discos: Esto se refiere a que se pueden utilizar varios discos que me permiten crear un drive logico, esto me indica que la informacion puede ser distribuida entre los discos y esto es bastante util para poder responder a los requerimientos de los usuarios de manera interna y externa (I/O).

Organization de Datos: Esto se debe a que estos discos utilizan varios metodos, como el metodo striping, donde la informacion es dividida y distribuida en bloques de datos que se encuentran en distintos discos, y el metodo mirroring, donde la informacion es duplicada y distribuida en distintos discos que se encuentran en multiples discos. Esto sirve para que la informacion pueda ser accedida de manera mas rapida y eficiente.

Redundancia: Esto se debe a que los discos en distintos niveles utilizan metodos como el mirroring, donde la informacion es duplicada y distribuida en distintos discos que se encuentran en multiples discos, esto permite guardar informacion de paridad o informacion duplicada de acuerdo al nivel de RAID que se este utilizando. Esto sirve para que la informacion pueda ser recuperada en caso de que algun disco falle.
